```python
import os
import random
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, History, ReduceLROnPlateau
from tensorflow.keras.optimizers import SGD
import matplotlib.pyplot as plt

# Setting seeds for reproducibility
seed = 42
random.seed(seed)
np.random.seed(seed)
tf.random.set_seed(seed)

# Load dataset from the CSV file
dataset = pd.read_csv('3D_printer.csv')

# Convert 'Material' column to numerical values
dataset['Material'] = dataset['Material'].replace({'abs': -1, 'pla': 1})

# One-hot encode 'infill_pattern' column
encoder = OneHotEncoder(sparse_output=False)
infill_pattern_encoded = encoder.fit_transform(dataset[['infill_pattern']])

# Separate features and labels
X = dataset.drop(columns=['Tension_strenght'])  # Features
y = dataset['Tension_strenght']  # Target variable (strength)

# Concatenate the one-hot encoded features with other numerical features
X = np.concatenate((X.drop(columns=['infill_pattern']).values, infill_pattern_encoded), axis=1)

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=seed)

# Define your Keras model creation function with variable number of neurons
def create_model(neurons=7, learning_rate=0.01, hidden_activation='selu', output_activation='linear'):
    model = Sequential()
    
    # Input layer
    model.add(Dense(2, input_shape=(X_train.shape[1],), activation=hidden_activation))
    
    # Adding fixed number of hidden layers with variable neurons
    for _ in range(4):  # 3 hidden layers
        model.add(Dense(neurons, activation=hidden_activation))
    
    # Output layer
    model.add(Dense(1, activation=output_activation))
    
    # Set the optimizer  with a specified learning rate
    opt = SGD(learning_rate=learning_rate)
    
    model.compile(optimizer=opt, loss='mean_absolute_error', metrics=['mean_absolute_error'])
    return model

# Hyperparameter tuning for neurons in the hidden layers
neuron_options = range(1, 21)  # From 1 to 20 neurons

# Initialize variables to track the best parameter
best_param_mae = float('inf')  # Initialize with infinity
best_param_diff = float('inf')  # Initialize with infinity
best_param_neurons = None
best_val_loss = float('inf')  # Initialize with infinity

for neurons in neuron_options:
    print(f"Training model with {neurons} neurons in hidden layers...")
    
    # Create History callback
    history = History()
    
    # Create the model with the current number of neurons
    model = create_model(neurons=neurons, learning_rate=0.01)  # Starting with a higher learning rate
    
    # Train the model with learning rate scheduler
    model.fit(X_train, y_train, epochs=1000, batch_size=16, validation_split=0.2, callbacks=[
        history,
        EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)  # Learning rate scheduler
    ], verbose=0)
    
    # Access the loss history
    loss_history = history.history['loss']
    val_loss_history = history.history['val_loss']
    
    # Find the lowest loss
    lowest_loss = min(loss_history)
    lowest_val_loss = min(val_loss_history)
    
    # Calculate the difference between training and validation loss
    train_val_diff = abs(lowest_loss - lowest_val_loss)
    
    # Calculate test MAE
    test_loss_mae = model.evaluate(X_test, y_test, verbose=0)[0]

    # Check if the current configuration has the least MAE
    if test_loss_mae < best_param_mae:
        best_param_mae = test_loss_mae
        best_param_diff = train_val_diff
        best_param_neurons = neurons
        best_val_loss = lowest_val_loss
    elif test_loss_mae == best_param_mae and train_val_diff < best_param_diff:
        # If the current configuration has the same MAE but smaller difference, update the best parameter
        best_param_diff = train_val_diff
        best_param_neurons = neurons
        best_val_loss = lowest_val_loss
    
    # Print the results
    print(f"Neurons: {neurons} - Lowest training loss: {lowest_loss}, Lowest validation loss: {lowest_val_loss}, Final test MAE: {test_loss_mae}, Train-Validation MAE Difference: {train_val_diff}")

    # Predict the values for the test set
    y_pred = model.predict(X_test).flatten()

    # Plot the real vs predicted values
    plt.figure(figsize=(10, 6))
    plt.plot(y_test.values, label='Real Values')
    plt.plot(y_pred, label='Predicted Values')
    plt.title(f'Real vs Predicted Values for {neurons} Neurons')
    plt.xlabel('Samples')
    plt.ylabel('Tension Strength')
    plt.legend()
    plt.show()

    # Scatter plot of real vs predicted values
    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, y_pred, alpha=0.5)
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')  # Diagonal line
    plt.title(f'Scatter Plot of Real vs Predicted Values for {neurons} Neurons')
    plt.xlabel('Real Values')
    plt.ylabel('Predicted Values')
    plt.show()

# Print the best parameter
print(f"Best Parameter - Neurons: {best_param_neurons}, Test MAE: {best_param_mae}, Validation Loss: {best_val_loss}, Train-Validation MAE Difference: {best_param_diff}")

# Retrain the best model with the entire training set
best_model = create_model(neurons=best_param_neurons, learning_rate=0.01)
best_model.fit(X_train, y_train, epochs=1000, batch_size=16, validation_split=0.2, callbacks=[
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)
], verbose=0)
```
